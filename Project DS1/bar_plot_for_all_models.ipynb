```python
df = pd.DataFrame(columns=["model_name","acc", "spec", "sens", "mode"],
                             data = data)
melted = pd.melt(df, id_vars=['model_name', "mode"], value_vars = ["acc", "sens", "spec"])
```

```python
plots = sns.barplot(data = melted[melted["mode"] == "multiclass"], 
                    x="model_name", y="value", hue = "variable")
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                (bar.get_x() + bar.get_width() / 2,
                bar.get_height()), ha='center', va='center',
                size=7, xytext=(0, 8),
                textcoords='offset points')
plt.title("Multiclass metrics")
plt.legend(bbox_to_anchor=(1, 1.05))
plt.show()
```

```python
plots = sns.barplot(data = melted[melted["mode"] == "binary"], 
                    x="model_name", y="value", hue = "variable")
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                (bar.get_x() + bar.get_width() / 2,
                bar.get_height()), ha='center', va='center',
                size=7, xytext=(0, 8),
                textcoords='offset points')
plt.title("Binary metrics")
plt.legend(bbox_to_anchor=(1, 1.05))
plt.show()
```
